```python
import os, shutil
import numpy as np
from zipfile import ZipFile
from zipfile import is_zipfile
```

Spacchettamento

```python
# Path alla directory che contiene tutti gli zip da spacchettare
directory = "/home/leonardo/Documents/DeepLearning/zipped_dataset"

# Destinazione dei file estratti
new_dir = "/home/leonardo/Documents/DeepLearning/Risiko_synthetic_dataset"

# Boh in caso sia un'altra l'estensione delle label
label_extension = ".txt"

# Creazione directory per immagini e label se non esiste gia'
new_images_dir = os.path.join(new_dir, "images")
new_labels_dir = os.path.join(new_dir, "labels")
if not os.path.isdir(new_images_dir): 
    os.mkdir(new_images_dir)
    print("Created images directory")
if not os.path.isdir(new_labels_dir): 
    os.mkdir(new_labels_dir)
    print("Created labels directory")


old_images_dir = os.path.join(directory, "synthetic_dataset/images")
old_labels_dir = os.path.join(directory, "synthetic_dataset/labels")

# Per rinominare i file
count = 0

for zipname in sorted(filter(lambda x: is_zipfile(os.path.join(directory, x)), os.listdir(directory))):
    
    # Estrazione dello zip
    zf = ZipFile(os.path.join(directory, zipname))
    zf.extractall(directory)
    print(zipname + " extracted")
    zf.close()

    # Rinominazione e spostamento di immagini e label
    for file in sorted(os.listdir(old_images_dir)):
        
        splitted_name = file.split(".")
        label = splitted_name[0] + label_extension
        new_img_name = str(count) + "." + splitted_name[1]
        new_lbl_name = str(count) + label_extension

        os.rename(os.path.join(old_images_dir, file), os.path.join(new_images_dir, new_img_name))
        print("Moved " + file + " and renamed into " + new_img_name)

        os.rename(os.path.join(old_labels_dir, label), os.path.join(new_labels_dir, new_lbl_name))
        print("Moved " + label + " and renamed into " + new_lbl_name)

        count += 1
    print(zipname + " -> done!")
    shutil.rmtree(directory + "/synthetic_dataset")
    print("Removed the old synthetic_dataset directory")
```

Crazione cartelle train-test-val

```python
dataset_dir = "/home/leonardo/Documents/DeepLearning/Risiko_synthetic_dataset"
train_test_dir = dataset_dir

# Creazione directory per train e test set
train_dir = os.path.join(train_test_dir, "train")
test_dir = os.path.join(train_test_dir, "test")
val_dir = os.path.join(train_test_dir, "val")
if not os.path.isdir(train_dir): 
    os.mkdir(train_dir)
    print("Created training directory")
if not os.path.isdir(test_dir): 
    os.mkdir(test_dir)
    print("Created test directory")
if not os.path.isdir(val_dir):
    os.mkdir(val_dir)
    print("Created validation directory")

# Creazione directory per immagini e label (train)
train_images_dir = os.path.join(train_test_dir, "train/images")
train_labels_dir = os.path.join(train_test_dir, "train/labels")
if not os.path.isdir(train_images_dir): 
    os.mkdir(train_images_dir)
    print("Created images directory (train)")
if not os.path.isdir(train_labels_dir): 
    os.mkdir(train_labels_dir)
    print("Created labels directory (train)")

# Creazione directory per immagini e label (train)
test_images_dir = os.path.join(train_test_dir, "test/images")
test_labels_dir = os.path.join(train_test_dir, "test/labels")
if not os.path.isdir(test_images_dir): 
    os.mkdir(test_images_dir)
    print("Created images directory (test)")
if not os.path.isdir(test_labels_dir): 
    os.mkdir(test_labels_dir)
    print("Created labels directory (test)")

# Creazione directory per immagini e label (validation)
val_images_dir = os.path.join(train_test_dir, "val/images")
val_labels_dir = os.path.join(train_test_dir, "val/labels")
if not os.path.isdir(val_images_dir): 
    os.mkdir(val_images_dir)
    print("Created images directory (validation)")
if not os.path.isdir(val_labels_dir): 
    os.mkdir(val_labels_dir)
    print("Created labels directory (validation)")
```

Divisione dataset

```python
# percentage of the dataset that will be used for test and validation set (over the total)
test_percentage = 30
val_percentage = 10

rng_test = np.random.default_rng(2023)
rng_val = np.random.default_rng(2023)

images_list = sorted(os.listdir(os.path.join(dataset_dir, "images")))
labels_list = sorted(os.listdir(os.path.join(dataset_dir, "labels")))

dataset_dim = len(images_list)

i = 0
while i < int(dataset_dim*(test_percentage/100)):
    rand_i = rng_test.integers(len(images_list))
    if os.path.isfile(dataset_dir + "/images/" + images_list[rand_i]):
        os.rename(dataset_dir + "/images/" + images_list[rand_i], os.path.join(test_images_dir, images_list[rand_i]))
        os.rename(dataset_dir + "/labels/" + labels_list[rand_i], os.path.join(test_labels_dir, labels_list[rand_i]))
        if i % 10 == 0: print(images_list[rand_i] + " selected for test")
        images_list.pop(rand_i)
        labels_list.pop(rand_i)
        i += 1
    else: i -= 1

i = 0
while i < int(dataset_dim*(val_percentage/100)):
    rand_i = rng_val.integers(len(images_list))
    if os.path.isfile(dataset_dir + "/images/" + images_list[rand_i]):
        os.rename(dataset_dir + "/images/" + images_list[rand_i], os.path.join(val_images_dir, images_list[rand_i]))
        os.rename(dataset_dir + "/labels/" + labels_list[rand_i], os.path.join(val_labels_dir, labels_list[rand_i]))
        if i % 10 == 0: print(images_list[rand_i] + " selected for validation")
        images_list.pop(rand_i)
        labels_list.pop(rand_i)
        i += 1
    else: i-= 1
        
#images_list = sorted(os.listdir(os.path.join(dataset_dir, "images")))
#labels_list = sorted(os.listdir(os.path.join(dataset_dir, "labels")))
for i in range(len(images_list)):
    os.rename(dataset_dir + "/images/" + images_list[i], os.path.join(train_images_dir, images_list[i]))
    os.rename(dataset_dir + "/labels/" + labels_list[i], os.path.join(train_labels_dir, labels_list[i]))

# rimozione vecchie cartelle images e labels
shutil.rmtree(dataset_dir + "/images")
shutil.rmtree(dataset_dir + "/labels")
print("Rimosse vecchie cartelle images e labels")
```
